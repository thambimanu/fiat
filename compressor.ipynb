```python
y=data1['bearings'] # copied to check
X=data1.drop('bearings',axis=1)# removed for model
```

```python
from sklearn.model_selection import train_test_split#.25,.30,.35
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35,random_state=0) #0.75 data will be for training.
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=1000)

#classifier=LogisticRegression()
classifier.fit(X_train,y_train) #command for training / fitting the model
```

```python
X_train.head(5)
```

```python
X_test.head(3)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
